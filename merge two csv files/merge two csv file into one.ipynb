# Objective :- In this project we are going to merge two csv files into one csv using python

```python
#Firstly we will import pandas, which we are going to use to merge two files
import pandas as pd
```

```python
# now we will import our dataset in a dataframe using pd.read_csv() function
student_detail = pd.read_csv('student detail.csv')
project_detail = pd.read_csv('project detail.csv')
```

 In the next step lets investigate what our data is
 for this we will use .head() function to see top rows of our dataset

```python
student_detail.head()
```

```python
project_detail.head()
```

This file contain project details assigned to students on the basis of roll no.

### Merging both the files

```python
# to merge the csv file we will use .merge() function
Msc_Python_Project = pd.merge(student_detail,project_detail,on='Roll No.')
```

```python
# our resultant dataframe will look like this
Msc_Python_Project
```

## Converting dataframe to csv file

```python
Msc_Python_Project.to_csv('Msc_python_project.csv',index=False)
```
